```python
import pybitflyer
import time

api_key = input("api_keyを入力し、Enterを押してください")
api_secret = input("api_secretを入力し、Enterを押してください")

api = pybitflyer.API(api_key=api_key, api_secret=api_secret)

switch_BTC = 1
switch_ETH = 0

#直近のBTC取引価格
base_price_BTC = api.ticker(product_code="BTC_JPY")["ltp"]
base_price_ETH = api.ticker(product_code="ETH_JPY")["ltp"]
# print("bot開始時の価格は"+str(base_price)+"です。")


while True:
	time.sleep(7)
	price_now_BTC = api.ticker(product_code="BTC_JPY")["ltp"]
	# print("現在のBTCJPYは "+str(price_now)+" です。")
	if base_price_BTC*0.999 >= price_now_BTC and switch_BTC == 1:
		print("BTC価格が下落しています。成行きで売り注文を入れます。")
		print("bitflyerのサーバーからの応答は以下です。")
		print(api.sendchildorder(product_code="BTC_JPY", child_order_type="MARKET",
		side="SELL", size=0.005, minute_to_expire=10000, time_in_force="GTC"))
		base_price_BTC = price_now_BTC
		switch_BTC = 0

	elif base_price_BTC*1.001 <= price_now_BTC and switch_BTC == 0:
		print("BTC価格が上昇しています。成行きで買い注文を入れます。")
		print("bitflyerのサーバーからの応答は以下です。")
		print(api.sendchildorder(product_code="BTC_JPY", child_order_type="MARKET",
		side="BUY", size=0.005, minute_to_expire=10000, time_in_force="GTC"))
		base_price_BTC = price_now_BTC
		switch_BTC = 1
		
	price_now_ETH = api.ticker(product_code="ETH_JPY")["ltp"]
	# print("現在のETHJPYは "+str(price_now)+" です。")
	if base_price_ETH*0.98 >= price_now_ETH and switch_ETH == 1:
		print("ETH価格が下落しています。成行きで売り注文を入れます。")
		print("bitflyerのサーバーからの応答は以下です。")
		print(api.sendchildorder(product_code="ETH_JPY", child_order_type="MARKET",
                           side="SELL", size=0.06, minute_to_expire=10000, time_in_force="GTC"))
		base_price_ETH = price_now_ETH
		switch_ETH = 0

	elif base_price_ETH*1.02 <= price_now_ETH and switch_ETH == 0:
		print("ETH価格が上昇しています。成行きで買い注文を入れます。")
		print("bitflyerのサーバーからの応答は以下です。")
		print(api.sendchildorder(product_code="ETH_JPY", child_order_type="MARKET",
                           side="BUY", size=0.06, minute_to_expire=10000, time_in_force="GTC"))
		base_price_ETH = price_now_ETH
		switch_ETH = 1
```
